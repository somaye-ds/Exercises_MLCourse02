```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
t_data = pd.read_csv('project-data.csv')
t_data
```

```python
#Delete unnecessery columns
t_data.drop(['<TICKER>','<PER>'], inplace = True, axis = 1)
t_data
```

```python
#Rename column names
t_data.columns.name = None
t_data.columns = ['Date','First_price','Higher_price', 'Lower_price', 'Close', 'Value', 'Vol', 'Openint','Open', 'Last']
t_data
```

```python
t_data['Date'] = pd.to_datetime(t_data['Date'], format='%Y%m%d')
```

```python
t_data
```

```python
t_data.info()
```

```python
Close = t_data.Close.astype(int)
Date = t_data.Date
```

```python
plt.plot(Date, Close)
```

```python
#Split "Close" list to group of 10
c_split = []
for i in range(0, len(t_data)+1, 10):
  c_list = list(Close[i:i+10])
  c_split.append(c_list)

c_split
```

```python
sets_predict = []
for i in range(0,len(c_split)-1):
  g = c_split[i]
  g_next = c_split[i+1]
  if g_next[0] > g[9]:
    sets_predict.append('True')
  else:
    sets_predict.append('False')
```
