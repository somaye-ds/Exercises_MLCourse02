```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/sample_data/california_housing_train.csv')
```

```python
df.head()
```

```python
house_val = df['median_house_value']
```

```python
df_x = df.drop(labels =['median_house_value'], axis = 1)
df_x
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
f1_sc_dict = {}
```

```python
for i in range(0,8):
  for j in range(i+1,8):
    df_x_new = df_x.drop(df.columns[[i,j]], axis=1)
    X_train, X_test, y_train, y_test = train_test_split(df_x_new, house_val)
    lr = LinearRegression()
    lr.fit(X_train, y_train)
    prediction = lr.predict(X_test)
    f1_sc_dict[i,j] = lr.score(X_test,y_test)
```

```python
f1_sc_dict
```

```python
max_f1_label = max(f1_sc_dict, key = f1_sc_dict.get) #tuple with len 2
min_f1_label = min(f1_sc_dict, key = f1_sc_dict.get) #tuple with len 2
```

```python
cl1 = df_x.columns[[max_f1_label[0],max_f1_label[1]]]
cl1
```

```python
cl2 = df_x.columns[[min_f1_label[0],min_f1_label[1]]]
cl2
```
